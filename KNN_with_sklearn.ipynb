## Predicting play will happen or not given 'Weather' and 'Temperature'  with KNN

### Creating our dataset

```
# First Feature
weather=['Sunny','Sunny','Overcast','Rainy','Rainy','Rainy','Overcast','Sunny','Sunny',
'Rainy','Sunny','Overcast','Overcast','Rainy']

# Second Feature
temp=['Hot','Hot','Hot','Mild','Cool','Cool','Cool','Mild','Cool','Mild','Mild','Mild','Hot','Mild']

# Label or target varible
play=['No','No','Yes','Yes','Yes','No','Yes','No','Yes','Yes','Yes','Yes','Yes','No']
```

### Encoding words to numbers

```
# Import LabelEncoder
from sklearn import preprocessing

#creating labelEncoder
le = preprocessing.LabelEncoder()

weather_encoded=le.fit_transform(weather)
print(weather_encoded)
```

```

temp_encoded=le.fit_transform(temp)
print(temp_encoded)

label=le.fit_transform(play)
print(temp_encoded)
```

```
#Combining Features
#Here, you will combine multiple columns or features into a single set of data using "zip" function

features=list(zip(weather_encoded,temp_encoded))
```

## Training

### Actually the KNN doesn't train on the training set unlike other ML models. It trains when a new datapoint is put for prediction.

```
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=3)
model.fit(features,label)
```

```
#Predict Output if weather is Overcast:1 and temp is Mild: 2
predicted= model.predict([[0, 2]]) 
print(predicted)

#Our model predicts the play will happen
```

Plotting a 3d scatter plot of our dataset

```
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt



fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

x = weather_encoded, 
y = temp_encoded,
z = label



ax.scatter(x, y, z, c= 'r', marker='o')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
